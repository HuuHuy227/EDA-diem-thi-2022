## Phổ điểm môn toán

```python
math_score = df['toan'].value_counts().sort_index()
print('Có {} thí sinh đạt điểm 0 môn toán'.format(math_score[0.0]))
print('Có {} thí sinh đạt điểm 10 môn toán'.format(math_score[10.0]))
```

```python
sns.set(style="darkgrid")
math_score.plot(kind="bar",figsize=(20,10),color=['#3CB371','#EE6363'])
plt.title('Phổ điểm môn toán',fontsize=15)
plt.xlabel('Điểm',fontsize=12)
plt.ylabel('Số thí sinh',fontsize=12)
plt.show()
```

## Phổ điểm môn văn

```python
liter_score = df['ngu_van'].value_counts().sort_index()
print('Có {} thí sinh đạt điểm 0 môn văn'.format(liter_score[0.0]))
print('Có {} thí sinh đạt điểm 10 môn văn'.format(liter_score[10.0]))
```

## Số thí sinh thi mỗi môn

```python
subjects = ['Toán','Văn','Tiếng Anh','Lý','Hóa','Sinh','Sử','Địa','GDCD']
count = []
for col in df.columns[1:]:
  selected_rows = df[~df[col].isna()]
  count.append(len(selected_rows))

# Plot chart
plt.figure(figsize=(15,7))
plt.barh(subjects, count,color=['#3CB371','#EE6363'])
for index, value in enumerate(count):
    plt.text(value, index,str(value))
plt.xlabel("Số thí sinh")
plt.title("Số thí sinh thi mỗi môn",fontsize=15)
plt.show()
```

## Điểm trung bình của từng môn học

```python
count1 = []
for col in df.columns[1:]:
  selected_rows = df[col].mean()
  count1.append(selected_rows)
dic = dict(zip(subjects,count1))
mean_data = pd.DataFrame(dic.items(), columns=['subject', 'count'])
# Plot chart
plt.figure(figsize=(15, 7))
plots = sns.barplot(x="subject", y="count", data=mean_data)
# Iterrating over the bars one-by-one
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')

plt.xlabel("Môn học", size=12)
plt.ylabel("Điểm trung bình", size=12)
plt.title("Điểm trung bình của từng môn học",fontsize=15)
plt.show()
```

## Số thí sinh thi ban KHTN so với ban KHXH

```python
khtn_rows = len(df[df[['vat_li','hoa_hoc','sinh_hoc']].notnull().all(axis=1)])
khxh_rows = len(df[df[['lich_su','dia_li','gdcd']].notnull().all(axis=1)])
y = [khtn_rows,khxh_rows]
x = ['Khoa học tự nhiên','Khoa học xã hội']
c = ['lightskyblue', 'thistle']
plt.figure(figsize=(15,7))
plt.title('Biểu đồ tròn giữa ban KHTN và KHXH', size=15)
plt.pie(y, colors=c)
plt.legend(x, loc='best', fontsize=12)
plt.show()
```

## Phân tích khối A01,A00,D01 trên thang điểm 40 với hệ số toán x2 

```python
df['D01'] = df['toan']*2 + df['ngu_van'] + df['ngoai_ngu']
df['A00'] = df['toan']*2 + df['vat_li'] + df['hoa_hoc']
df['A01'] = df['toan']*2 + df['vat_li'] + df['ngoai_ngu']
df.head(10)
```

```python
ranges = [0,5,10,15,20,25,30,35,40]
d01_count = df['D01'].groupby(pd.cut(df.D01, ranges)).count()

d01_count.plot(kind="bar",figsize=(20,10),color=['#3CB371','#EE6363'])
plt.title('Phổ điểm khối D01 trên thang 40',fontsize=15)
plt.xlabel('Khoảng điểm',fontsize=12)
plt.ylabel('Số thí sinh',fontsize=12)
plt.show()
```

```python
ranges = [0,5,10,15,20,25,30,35,40]
a00_count = df['A00'].groupby(pd.cut(df.A00, ranges)).count()

a00_count.plot(kind="bar",figsize=(20,10),color=['#3CB371','#EE6363'])
plt.title('Phổ điểm khối A00 trên thang 40',fontsize=15)
plt.xlabel('Khoảng điểm',fontsize=12)
plt.ylabel('Số thí sinh',fontsize=12)
plt.show()
```
